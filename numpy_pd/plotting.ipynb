## Images

```python
import urllib.request

# Setting custom User-Agent header to avoid 429 rate limit error on imgur
opener = urllib.request.build_opener()
opener.addheaders = [
    (
        "User-agent",
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
    )
]
urllib.request.install_opener(opener)
urllib.request.urlretrieve("https://i.imgur.com/SkPbq.jpeg", "chart.jpg")

from PIL import Image

img = Image.open('chart.jpg')
img
```

```python
img_array = np.array(img)
img_array.shape
```

```python
# Convert to grayscale (need to set dtype to uint8 because color range for float is 0.0 - 1.0 but we have 0-255)
gray_scale_img = np.zeros(img_array.shape, dtype=np.uint8)
for i in range(img_array.shape[0]):
    for j in range(img_array.shape[1]):
        avg = (img_array[i, j, 0] + img_array[i, j, 1] + img_array[i, j, 2]) / 3
        for k in range(3):
            gray_scale_img[i, j, k] = avg
```

```python
plt.imshow(gray_scale_img)
```

```python
plt.grid(False)
plt.title("A data science meme")
plt.axis('off')
plt.imshow(img)
```
